```python
# data analysis
import pandas as pd

# plot tool
import matplotlib.pyplot as plt

# split train and test data
from sklearn.model_selection import train_test_split

# scale the data points
from sklearn.preprocessing import StandardScaler 

# Neural Network model
from sklearn.neural_network import MLPClassifier

# Confusion marix to evaluate the model
from sklearn.metrics import classification_report,confusion_matrix

# Evalute the speed of model
import datetime

# Exportation of  model
import pickle

# read data
raw_data = pd.read_csv(r'C:\Users\User\Desktop\Study\Year 3\Sem 1\CG3002\dataset.csv')

cleaned_data = raw_data.drop(['date', 'time', 'username', 'wrist'], axis = 1)

mean_data = pd.DataFrame()
std_data = pd.DataFrame()
i = 0
j = 99
row_count = len(cleaned_data.index)

# segmentation at 50%
while row_count > 1:
    a = cleaned_data[i:i+200:].copy()
    b = cleaned_data[j:j+200:].copy()
    
    a.loc['mean'] = a.mean()
    a.loc['std'] = a.std()
    b.loc['mean'] = b.mean()
    b.loc['std'] = b.std()
    mean_data = mean_data.append(a.loc['mean'], ignore_index = True)
    mean_data = mean_data.append(b.loc['mean'], ignore_index = True)
    std_data = std_data.append(a.loc['std'], ignore_index = True)
    std_data = std_data.append(b.loc['std'], ignore_index = True)
    
    i += 200
    j += 200
    row_count -= 200
    
mean_data.loc[mean_data['activity'] >= 0.5, 'activity'] = 1
mean_data.loc[mean_data['activity'] < 0.5, 'activity'] = 0

# rearrangement of columns
mean_data = mean_data.rename(index=str, columns={'acceleration_x': 'mean_acc_x', 'acceleration_y': 'mean_acc_y',
                                                 'acceleration_z': 'mean_acc_z', 'gyro_x': 'mean_gyro_x',
                                                 'gyro_y': 'mean_gyro_y', 'gyro_z': 'mean_gyro_z'})

std_data = std_data.rename(index=str, columns={'acceleration_x': 'std_acc_x', 'acceleration_y': 'std_acc_y',
                                                 'acceleration_z': 'std_acc_z', 'gyro_x': 'std_gyro_x',
                                                 'gyro_y': 'std_gyro_y', 'gyro_z': 'std_gyro_z'})
# combine features extracted into 1 
activity = mean_data['activity'].copy()
mean_data = mean_data.drop(['activity'], axis = 1)
std_data = std_data.drop(['activity'], axis = 1)
extracted_data = mean_data.join(std_data)
extracted_data['activity'] = activity

X = extracted_data.drop(['activity'], axis = 1)
y = extracted_data['activity'].copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)

scaler = StandardScaler()
# Fit only to the training data
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

start_time = datetime.datetime.now()

mlp = MLPClassifier(hidden_layer_sizes = 100)
mlp.fit(X_train, y_train)

end_time = datetime.datetime.now()
print("The training accuracy for Neural Network is ", mlp.score(X_train, y_train))
print("The testing accuracy for Neural Network is ", mlp.score(X_test, y_test))
time_taken = end_time - start_time
print("time taken = ", time_taken)

predictions = mlp.predict(X_test)

cm = confusion_matrix(y_test, predictions)

print(cm)


print(classification_report(y_test,predictions))

start_time = datetime.datetime.now()
# export model
filename = "Neural_Network_Finalized_Model.sav"
pickle.dump(mlp, open(filename, 'wb'))

loaded_model = pickle.load(open(filename, 'rb'))
print("The testing accuracy for imported Neural Network is ", loaded_model.score(X_test, y_test))
predictions = loaded_model.predict(X_test)
print(predictions)

end_time = datetime.datetime.now()
time_taken = end_time - start_time
print("time taken = ", time_taken)

extracted_data.to_csv(r'C:\Users\User\Desktop\Study\Year 3\Sem 1\CG3002\cleaned_dataset.csv',index=False)
```
