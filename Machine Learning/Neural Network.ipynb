```python
# data analysis
import pandas as pd

# plot tool
import matplotlib.pyplot as plt

# split train and test data
from sklearn.model_selection import train_test_split

# scale the data points
from sklearn.preprocessing import StandardScaler 

# Neural Network model
from sklearn.neural_network import MLPClassifier

# Confusion marix to evaluate the model
from sklearn.metrics import classification_report,confusion_matrix

# Evalute the speed of model
import datetime

# Exportation of  model
import pickle

# read data
raw_data = pd.read_csv(r'C:\Users\User\Desktop\Study\Year 3\Sem 1\CG3002\dataset.csv')

cleaned_data = raw_data.drop(['date', 'time', 'username'], axis = 1)

X = cleaned_data.drop(['activity'], axis = 1)
y = cleaned_data['activity']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)

scaler = StandardScaler()
# Fit only to the training data
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

a = datetime.datetime.now()

mlp = MLPClassifier(hidden_layer_sizes = 100)
mlp.fit(X_train, y_train)

b = datetime.datetime.now()
print("The training accuracy for Neural Network is ", mlp.score(X_train, y_train))
print("The testing accuracy for Neural Network is ", mlp.score(X_test, y_test))
c = b - a
print("time taken = ", c)

predictions = mlp.predict(X_test)

cm = confusion_matrix(y_test, predictions)

print(cm)


print(classification_report(y_test,predictions))

a = datetime.datetime.now()
# export model
filename = "Neural_Network_Finalized_Model.sav"
pickle.dump(mlp, open(filename, 'wb'))

loaded_model = pickle.load(open(filename, 'rb'))
print("The testing accuracy for imported Neural Network is ", loaded_model.score(X_test, y_test))
predictions = loaded_model.predict(X_test)
print(predictions)

b = datetime.datetime.now()
c = b - a
print("time taken = ", c)

cleaned_data.to_csv(r'C:\Users\User\Desktop\Study\Year 3\Sem 1\CG3002\cleaned_dataset.csv',index=False)
print(raw_data.head())
print(cleaned_data.head())
```
